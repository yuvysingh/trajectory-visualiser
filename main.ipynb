```python
import matplotlib.pyplot as plt 
import numpy as np
import math 
from matplotlib.animation import FuncAnimation
%matplotlib inline
```

```python
import cmath
def quadratic_solver(a,b,c):
    
    d = (b**2) - (4*a*c)
    pos_sol = (-b-cmath.sqrt(d))/(2*a)
    return pos_sol
```

```python
class Projectile:
    def __init__(self,initial_x,initial_y,intial_speed,angle_of_projection):
        self.g = 9.8
        self.x = initial_x
        self.y = initial_y
        self.u = intial_speed
        self.theta = angle_of_projection * (math.pi/180)
        self.time_of_flight = quadratic_solver((-self.g)/2,self.u * math.sin(self.theta),self.y)
        
    
    def path_visualiser(self):
        t= np.linspace(0,self.time_of_flight)
        h = (t * math.cos(self.theta) * self.u) + self.x
        v = (self.u * math.sin(self.theta) * t) - (((self.g)/2) * (t ** 2)) + self.y

        plt.plot(h,v)
        plt.xlabel("horizontal displacement")
        plt.ylabel("vertical displacement")
        plt.gridline()
        plt.title("projectile")
    
    
        
```

```python
ball = Projectile(0,0,5,30)
ball.path_visualiser()

animation = FuncAnimation()

fig = plt.figure()
```
